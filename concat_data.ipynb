```python
import json
import pandas as pd
```

```python
pdf_main = pd.read_csv('生成ファイル/AIICO_ポリゴンリスト_ユニットID付き.csv', header=0, index_col=False)
pdf_main = pdf_main[['caption', 'address']].drop_duplicates()

pdf_01     = pd.read_parquet('生成ファイル/遊戯王_separate.parquet')
pdf_02     = pd.read_parquet('生成ファイル/遊戯王_analysis.parquet')
pdf_02['分析結果'] = pdf_02['分析結果'].apply(json.loads).apply(lambda x: x.get('text'))
# pdf_02['分析結果'] = pdf_02['分析結果'].apply(json.loads).apply(lambda x: x.get('text')).str.replace(r'[\n\r,、]|\\n', '', regex=True)
pdf_concat = pdf_main\
				.merge(pdf_01, how='inner', on='caption')\
                .merge(pdf_02, how='inner', on='caption')
pdf_concat.to_csv('生成ファイル/aiico_cohort_遊戯王.csv', header=True, index=False)

pdf_03     = pd.read_parquet('生成ファイル/ポケモンカード_separate.parquet')
pdf_04     = pd.read_parquet('生成ファイル/ポケモンカード_analysis.parquet')
pdf_04['分析結果'] = pdf_04['分析結果'].apply(json.loads).apply(lambda x: x.get('text'))
# pdf_04['分析結果'] = pdf_04['分析結果'].apply(json.loads).apply(lambda x: x.get('text')).str.replace(r'[\n\r,、]|\\n', '', regex=True)
pdf_concat = pdf_main\
				.merge(pdf_03, how='inner', on='caption')\
                .merge(pdf_04, how='inner', on='caption')
pdf_concat.to_csv('生成ファイル/aiico_cohort_ポケモンカード.csv', header=True, index=False)

pdf_concat
```
